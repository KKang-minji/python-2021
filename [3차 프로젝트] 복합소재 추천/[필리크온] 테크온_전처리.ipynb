```python
PC_c = ta["수지종류"].str.contains("PC",na=False).value_counts()
PBT_c = ta["수지종류"].str.contains("PBT",na=False).value_counts()
TPC_c = ta["수지종류"].str.contains("TPC",na=False).value_counts()
ABS = ta["수지종류"].str.contains("ABS",na=False).value_counts()
PPS = ta["수지종류"].str.contains("PPS",na=False).value_counts()
PA = ta["수지종류"].str.contains("PA",na=False).value_counts()
POM= ta["수지종류"].str.contains("POM",na=False).value_counts()
PP = ta["수지종류"].str.contains("PP",na=False).value_counts()
PE = ta["수지종류"].str.contains("PE",na=False).value_counts()
PC_c, PBT_c, TPC_c ,PPS, ABS, PA, POM, PP, PE
```

```python
ta_tmp1 = ta.replace("TPC-ET",".")
ta_tmp1
ta_tmp2 = ta.replace("PPA/GF",".")
ta_tmp2
```

```python
ta_tmp3=ta.replace("FR PP","PP").replace("PP+GF20%","PP").replace("PP+GF30%","PP").replace("PP+GF45%","PP").replace("PP+MF20%","PP").replace("PP+MF30%","PP").replace("PP+MF30%","PP").replace("PP+MF40%","PP").replace("PP+Rubber+MF","PP").replace("PP+GF/MF40%","PP")
```

```python
ta_1=ta.loc[ta['수지종류'].str.contains('ABS', na=False)]
ta_2=ta.loc[ta['수지종류'].str.contains('PBT', na=False)]
ta_3=ta.loc[ta['수지종류'].str.contains('PPS', na=False)]
ta_4=ta_tmp2.loc[ta_tmp2['수지종류'].str.contains('PA', na=False)]
ta_5=ta.loc[ta['수지종류'].str.contains('POM', na=False)]
ta_6=ta_tmp3.loc[ta_tmp3['수지종류']=="PP"]
ta_7=ta.loc[ta['수지종류'].str.contains('PE', na=False)]
ta_8=ta.loc[ta['수지종류'].str.contains('TPC', na=False)]
ta_9=ta.loc[ta['수지종류'].str.contains('PPA', na=False)]
ta_10=ta_tmp1.loc[ta_tmp1['수지종류'].str.contains('PC', na=False)]
pd.concat([ta_1,ta_2,ta_3,ta_4,ta_5,ta_6,ta_7,ta_8,ta_9,ta_10])
```

```python
ta_1.loc[ta_1["수지종류"].str.contains(pat="ABS"), "수지종류"] = "ABS"
ta_2.loc[ta_2["수지종류"].str.contains(pat="PBT"), "수지종류"] = "PBT"
ta_3.loc[ta_3["수지종류"].str.contains(pat="PPS"), "수지종류"] = "PPS"
ta_4.loc[ta_4["수지종류"].str.contains(pat="PA"), "수지종류"] = "PA"
ta_5.loc[ta_5["수지종류"].str.contains(pat="POM"), "수지종류"] = "POM"
ta_7.loc[ta_7["수지종류"].str.contains(pat="PE"), "수지종류"] = "PE"
ta_8.loc[ta_8["수지종류"].str.contains(pat="TPC"), "수지종류"] = "TPC"
ta_9.loc[ta_9["수지종류"].str.contains(pat="PPA"), "수지종류"] = "PPA"
ta_10.loc[ta_10["수지종류"].str.contains(pat="PC"), "수지종류"] = "PC"
taek_1 = pd.concat([ta_1,ta_2,ta_3,ta_4,ta_5,ta_6,ta_7,ta_8,ta_9,ta_10])
taek_1
```
