```python
import pandas as pd # DataFrame 을 위한 모듈
import numpy as np # arrange 
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Gulim' 
```

```python
ta = pd.read_csv("테크온_전처리.csv", encoding="cp949", index_col=0)
ta
```

### 테크온 기본 물성표

```python
ta2=pd.DataFrame({'구분':['EFRP', 'SFRP', 'CFRP(고강도)', 'CFRP(고탄성)'],
              '비중':['2.55', '2.5', '1.8', '1.98'],
              '인장탄성율':['3400', '4480', '6500', '3400'],
             '인장강도':['72', '87', '300', '650'],
             '신율':['4.8', '5.7', '2.1', '0.5']})
ta2
```

### 수지종류별 "비중(Specific Gravity)" 그래프와 기본 물성표 비교

```python
sns.boxplot(x="수지종류", y="비중(Specific Gravity)", data=ta)
line1 = plt.axhline(y=2.55, color='y', label='EFRP', linewidth=1)
line2 = plt.axhline(y=2.5, color='g', label='SFRP', linewidth=1)
line3 = plt.axhline(y=1.8, color='r', label='CFRP(고강도)', linewidth=1.5)
line4 = plt.axhline(y=1.98, color='b', label='CFRP(고탄성)', linewidth=1.5)
plt.legend(handles=[line1, line2, line3, line4])
plt.show()

#frp는 데이터에 비해 수치가 매우높은 편이고 cfrp는 pps와 겹친다.
```

### 수지종류별 "인장강도(Mpa)" 그래프와 기본 물성표 비교

```python
sns.boxplot(x="수지종류", y="인장탄성율(Mpa)", data=ta)
line1 = plt.axhline(y=3400, color='y', label='EFRP', linewidth=1)
line2 = plt.axhline(y=4480, color='g', label='SFRP', linewidth=1)
line3 = plt.axhline(y=6500, color='r', label='CFRP(고강도)', linewidth=1.5)
line4 = plt.axhline(y=3400, color='b', label='CFRP(고탄성)', linewidth=1.5)
plt.legend(handles=[line1, line2, line3, line4])

plt.show()

#모든 기준선들이 abs,pbt,pa,pc와 겹치고, 특히 pbt,pa가 가장 많이 겹친다.
```

### 수지종류별 "신율(%)" 그래프와 기본 물성표 비교

```python
sns.boxplot(x="수지종류", y="신율\n(%)", data=ta)
line1 = plt.axhline(y=4.8, color='y', label='EFRP', linewidth=1.5)
line2 = plt.axhline(y=5.7, color='g', label='SFRP', linewidth=1.5)
line3 = plt.axhline(y=2.1, color='r', label='CFRP(고강도)', linewidth=1.5)
line4 = plt.axhline(y=0.5, color='b', label='CFRP(고탄성)', linewidth=1.5)
plt.legend(handles=[line1, line2, line3, line4])
plt.show()

#신율 기준선이 잘 보이지않아서 선가까이에 있는 것빼고 다 drop
```

```python
### drop 후 어떤 수지와 기본물성표 수치와 가까운지 비교
```

```python
ta=ta[ta.수지종류 != 'ABS']
ta=ta[ta.수지종류 != 'PBT']
ta=ta[ta.수지종류 != 'PA']
ta=ta[ta.수지종류 != 'PP']
ta=ta[ta.수지종류 != 'POM']
ta=ta[ta.수지종류 != 'PC']
ta=ta[ta.수지종류 != 'TPC']
```

```python
sns.boxplot(x="수지종류", y="신율\n(%)", data=ta)
line1 = plt.axhline(y=4.8, color='y', label='EFRP', linewidth=1.5)
line2 = plt.axhline(y=5.7, color='g', label='SFRP', linewidth=1.5)
line3 = plt.axhline(y=2.1, color='r', label='CFRP(고강도)', linewidth=1.5)
line4 = plt.axhline(y=0.5, color='b', label='CFRP(고탄성)', linewidth=1.5)
plt.legend(handles=[line1, line2, line3, line4])
plt.show()


#테크온에서 원하는 FRP수치들과 PE수지수치가 비슷하다
#PPS수지에 테크온에서 원하는 CFRP수치가 있을 확률이 높다?
```

# 결측 해결 후 시각화

```python
ta1 = pd.read_csv("테크온_결측치평균.csv",encoding="utf-8")
ta1
```

```python
ta1= ta1.rename(columns={'세부':'구분','비중(Specific Gravity)':'비중','신율\n(%)':'신율','인장탄성율(Mpa)':'인장탄성율'}) 
```

```python
ta_1=ta1.loc[:,['구분','비중','인장강도','인장탄성율','신율']]
ta_1
```

```python
ta2=pd.DataFrame({'구분':['EFRP', 'SFRP', 'CFRP(고강도)', 'CFRP(고탄성)'],
              '비중':['2.55', '2.5', '1.8', '1.98'],
              '인장탄성율':['3400', '4480', '6500', '3400'],
             '인장강도':['72', '87', '300', '650'],
             '신율':['4.8', '5.7', '2.1', '0.5']})
ta2
```

```python
ta3 = pd.concat([ta_1,ta2])
ta3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#비중과 인장강고
sns.scatterplot(x='비중', y='인장강도', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['비중'],y=ta2['인장강도'],hue=ta2['구분'],style=ta2['구분'],s=200)

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()


#비중과 인장탄성율
sns.scatterplot(x='비중', y='인장탄성율', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['비중'],y=ta2['인장탄성율'],hue=ta2['구분'],style=ta2['구분'],s=200)


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()


#비중과 신율
sns.scatterplot(x='비중', y='신율', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['비중'],y=ta2['신율'],hue=ta2['구분'],style=ta2['구분'],s=200)


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()


#인장강도와 인장탄성율
sns.scatterplot(x='인장강도', y='인장탄성율', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['인장강도'],y=ta2['인장탄성율'],hue=ta2['구분'],style=ta2['구분'],s=200)


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()


#인장강도와 신율
sns.scatterplot(x='인장강도', y='신율', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['인장강도'],y=ta2['신율'],hue=ta2['구분'],style=ta2['구분'],s=200)


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()

#인장탄성율과 신ㅇ율
sns.scatterplot(x='인장탄성율', y='신율', hue='구분', data=ta_1)
sns.scatterplot(x=ta2['인장탄성율'],y=ta2['신율'],hue=ta2['구분'],style=ta2['구분'],s=200)


plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
