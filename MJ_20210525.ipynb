```python
#피타고라스 정리 1~100
count=0
for a in range(1,101):
    for b in range(1,101):
        for c in range(1,101):
            if a*a+b*b==c*c:  
                print(a,b,c)  #조합하는 a,b,c 추출
                count+=1
print("총 조합개수는", count, "개 입니다.")   #총 count  #if가 끝나는 곳에 print
```

```python
# 1 ~ 999 자연수 중 각자리에 있는 숫자가 서로 다른 자연수 모두 몇개 인가?
count = 0
for a in range(0, 1000):
    d100 = a//100
    d10 = (a%100)//10
    d1 = a%10
    if d100 != d10 and d100 != d1 and d10 != d1:  #양쪽 비교 연산자 여서 양쪽비교
        count+= 1
        print(d100,d10,d1)
print("자연수는",count,"개 입니다,", end="")
```

```python
def add_many(*a):
    result = 0
    for item in a:  # a가 (1, 2, 3, 4)랑 같은 얘기
        result += item
    return result
print(add_many(1, 2))
print(add_many(1, 2, 3, 4))
```

```python
def add_many(mode,*a):  #*a: 뭉텅이로 지정, list처리
    if mode =="sum":
        result=0
        for item in a:
            result+=item
    elif mode == "mul":
        result = 1
        for item in a: result+=item
    return result
print(add_many("sum",1,2))
print(add_many("mul",1,2,3,4))
```
